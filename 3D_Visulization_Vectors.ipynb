```python
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(0.5, 0.2, 0.1, color='r')

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Draw an arrow from origin to (0.5, 0.5, 0.5)
ax.quiver(0, 0, 0, 0.5, 0.5, 0.5, color='g', arrow_length_ratio=0.1)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
vectors = [(0.8, 0.2, 0.5), (0.3, 0.9, 0.4), (0.6, 0.6, 0.1)]
colors = ['r', 'g', 'b']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for (x, y, z, color) in zip(*zip(*vectors), colors):
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.1)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
plt.show()
```

```python
vectors = [(0.7, 0.2, 0.6), (0.4, 0.8, 0.3)]
words = ['apple', 'banana']
colors = ['c', 'm']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for (x, y, z, word, color) in zip(*zip(*vectors), words, colors):
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.1)
    ax.text(x, y, z, word, fontsize=10, color=color)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
plt.show()
```

```python
# Complete vector plot with annotations
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_coords = [0.7, 0.2, 0.5]
y_coords = [0.3, 0.9, 0.4]
z_coords = [0.6, 0.1, 0.8]
words = ['king', 'queen', 'man']
colors = ['r', 'g', 'b']

for (x, y, z, word, color) in zip(x_coords, y_coords, z_coords, words, colors):
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.05)
    ax.text(x, y, z, word, fontsize=10, color=color)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
plt.title('3D Plot of Word Embeddings')
plt.show()
```
